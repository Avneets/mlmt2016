```python
%matplotlib notebook
import pylab
pylab.rcParams['figure.figsize'] = (10., 10.)

import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 25
w_0, w_1 = 1, 0.35
sigma = 0.05
x = np.random.rand(N)
y = w_0 + w_1 * x + np.random.randn(N) * sigma
xx = np.linspace(0, 1, 100)
ytrue = w_0 + w_1 * xx

def plotFigure():
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, y, ls='', marker='x', markersize=8, markeredgewidth=2, label='Data')
    ax.plot(x, y - w_0, ls='', marker='x', markersize=8, markeredgewidth=2, label='Data')
    ax.plot(xx, ytrue, c='k', lw=2)
    ax.plot(xx, ytrue - w_0, c='k', lw=2)



plotFigure()
```

```python
N = 50
D = 2
sigma1, sigma2 = 3, 1
mu = np.array([8, 10])
rotangle = np.pi/5
rotmat = np.array([[np.cos(rotangle), np.sin(rotangle)], [-np.sin(rotangle), np.cos(rotangle)]])

X = np.random.randn(N, 2)
X[:, 0] *= sigma1
X[:, 1] *= sigma2

X = np.dot(X, rotmat)
X = X + mu.reshape(1, D)

def plotFigure2():
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xlim(-10, 15)
    ax.set_ylim(-10, 15)
    plt.plot([0, 0], [-10, 15], c='k', ls=':')
    plt.plot([-10, 15], [0, 0],  c='k', ls=':')
    
    ax.scatter(X[:, 0], X[:, 1], marker='x', lw=2, c='k', s=30, label='Original Inputs')
    
    X2 = X - X.mean(axis=0)
    X2 /= X2.std(axis=0)
    ax.scatter(X2[:, 0], X2[:, 1], marker='+', lw=2, c='b', s=30, label='Standardised Inputs')
    
    ax.legend(fontsize=25, loc='lower right')


    



plotFigure2()
X.mean(axis=0)
```
