```python
%matplotlib notebook
import pylab
pylab.rcParams['figure.figsize'] = (10., 10.)

from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from sklearn import linear_model

import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
# Let's plot the coin toss example with three observed tails TTT

def beta(thetas, a, b):
    return thetas**(a - 1) * (1 - thetas)**(b - 1) / sp.beta(a, b)

def plotBayes(thetas, prior, posterior):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    
    ax.plot(thetas, prior, lw='2', label='Prior')
    ax.plot(thetas, posterior, lw='2', label='Posterior')
    
    ax.legend(loc='upper right', fontsize=25)
    ax.tick_params(axis='both', labelsize=20)

nheads = 0
ntails = 3
thetas = np.linspace(0, 1, 100)
unifprior = np.ones(thetas.shape)
unifposterior = beta(thetas, nheads + 1, ntails + 1)

beta22prior = beta(thetas, 2, 2)
beta22posterior = beta(thetas, nheads + 2, ntails + 2)
```

```python
plotBayes(thetas, unifprior, unifposterior)
#plotBayes(thetas, beta22prior, beta22posterior)
```
