```python
%matplotlib notebook
from matplotlib import pylab
pylab.rcParams['figure.figsize'] = (12., 8.)

from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.ndimage import imread
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits

import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
alan = imread('alan.jpg', mode='F')
rec_errs = []
K_vals = [2, 5, 20, 50, 100, 300]
for (i, k) in enumerate(K_vals):
    pca = PCA(n_components=k)
    pca.fit(alan)
    Z = pca.fit_transform(alan)
    X = pca.inverse_transform(Z)
    ax = fig.add_subplot(2, 3, i + 1)
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_xlabel('K = %d' % k, fontsize=20)
    ax.imshow(X, cmap='Greys_r')
    rec_errs.append(np.linalg.norm(X - alan, ord='fro'))
```

```python
rec_errs = []
K_vals = [20, 25, 40, 60, 100, 150, 200, 300]
for (i, k) in enumerate(K_vals):
    pca = PCA(n_components=k)
    pca.fit(alan)
    Z = pca.fit_transform(alan)
    X = pca.inverse_transform(Z)
    rec_errs.append(np.linalg.norm(X - alan, ord='fro'))
```

```python
pylab.rcParams['figure.figsize'] = (10., 10.)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.tick_params(axis='both', labelsize=20)
ax.set_xlabel('# Principal Components Used', fontsize=20)
ax.set_ylabel('Reconstruction Error', fontsize=20)

ax.plot(K_vals, rec_errs/np.linalg.norm(alan, ord='fro'), marker='x', markersize=10, markeredgewidth=3, c='b', lw='2', label='Reconstruction Error')
```
