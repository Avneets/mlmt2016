```python
%matplotlib notebook

from matplotlib import pylab
pylab.rcParams['figure.figsize'] = (10.0, 10.0)

from matplotlib.patches import Arrow, Circle
from sklearn import svm

import matplotlib.pyplot as plt
import numpy as np
```

```python
fig  = plt.figure()
ax = fig.add_subplot(1, 1, 1)

xmin, xmax = -5, 5
ymin, ymax = -2.5, 5
ax.set_aspect('equal')
ax.set_xlim((xmin, xmax))
ax.set_ylim((ymin, ymax))

# Axes
ax.plot([xmin, xmax], [0, 0], ls=':', lw=1, c='k')
ax.plot([0, 0], [ymin, ymax], ls=':', lw=1, c='k')

# Line 2x + 3y = 6
x0, x1 = -2.5, 4.5
ax.plot([x0, x1], [(6 - 2*x0)/3., (6 - 2*x1)/3.], lw=2)
ax.arrow(0, 0, 2/np.sqrt(13), 3/np.sqrt(13), head_length=0.1, head_width =0.1, fc='k', ec='k', lw=2)
ax.plot([4], [2], ls='', marker='o', markeredgecolor='k', markerfacecolor='k', markersize=6)
ax.plot([0, 4], [0, 2], ls=':', c='k')
ax.plot([0, 28./13], [0, 42./13], ls=':', c='k')
ax.plot([28./13, 4], [42./13, 2], ls=':', c='k')
ax.plot([4 - 16./13, 4], [2 - 24./13, 2], c='m', lw=2)

alpha=0.06
beta = 0.03
ax.plot([-3 * alpha, -3 * alpha + 12./13], [2 * alpha, 2*alpha + 18./13])
ax.plot([-3 * beta, -3 * beta + 28./13], [2 * beta, 2*beta + 42./13])

#  ax.annotate(s='$-w_0$', xy=(1,1), xytext=(0,0), arrowprops=dict(arrowstyle='<->'))

ax.text(0.05, 1.05, '$-w_0/\Vert \mathbf{w} \Vert_2$',
        horizontalalignment='center',
        verticalalignment='center',
        rotation=60, fontsize=18, color='r')
ax.text(1.4, 2.6, '$\mathbf{w} \cdot \mathbf{x}/\Vert \mathbf{w} \Vert_2$',
        horizontalalignment='center',
        verticalalignment='center',
        rotation=60, fontsize=18, color='c')
ax.annotate(s='$(\mathbf{w} \cdot \mathbf{x} + w_0)/\Vert \mathbf{w}\Vert_2$', xy=(3.2, 0.8), xytext=(1.5, -1.5), arrowprops=dict(arrowstyle='->'), fontsize=25)
```
