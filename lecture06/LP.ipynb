```python
%matplotlib notebook
import pylab
pylab.rcParams['figure.figsize'] = (10., 10.)

from matplotlib.patches import Polygon
from matplotlib.font_manager import FontProperties

import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# l1: y = 5/2 - 3x/4
# l2: y = x/3 + 4
# l3: x = 4
# l4: y = 4x + 12

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_xticks(())
ax.set_yticks(())

ax.plot([0, 0], [-10, 10], c='k', ls=':', lw=0.8)
ax.plot([-10, 10], [0, 0], c='k', ls=':', lw=0.8)


ax.plot([-1.5, 4.2], [29./8, -0.65], c='c', lw=2) # l1
ax.plot([-2.5, -1], [19./6, 11./3], c='c', lw=2) #l2
ax.plot([4, 4], [-10, 0.], c='c', lw=2)
ax.plot([-10, -2.1], [-28, 3.6], c='c', lw=2)
poly_pts = [[-5.5, -10], [-24./11, 36./11], [-18./13, 46./13], [4., -.5], [4, -10]]
ax.add_patch(Polygon(poly_pts, fill=False, hatch='/', closed=True))
ax.arrow(0, 0, -4.5, -3, fc='m', lw=3, ec='m', head_width=0.3, head_length=0.8)
font = FontProperties()
font.set_weight('bold')
ax.text(-5.65, -4.1, 'c', fontsize=20, color='m', fontproperties=font)

ax.plot([4], [-.5], c='r', marker='o', markersize=10)
ax.text(4.2, -.5, 'Optimum', fontsize=20)
```
