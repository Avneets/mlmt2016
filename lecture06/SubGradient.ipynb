```python
%matplotlib notebook
import pylab
pylab.rcParams['figure.figsize'] = (10., 10.)

from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d

import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 0.1 * x**2

def f_1(x):
    return 0.2 * x

def g(x, a=2):
    if x < a:
        return f(x)
    else:
        return 2 * (x - a) + f(a)
gg = np.vectorize(g)
```

```python
xx = np.linspace(-4, 4, 200)
yy1 = f(xx)
yy2 = gg(xx)

x_0 = 2
y_0 = f(x_0)

yy1_deriv = (xx - x_0) * f_1(x_0) + f(x_0)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_ylim(-0.5, 2.5)
ax.set_xlim(-4, 4)

ax.tick_params(axis='both', labelsize=20)

plt.plot(xx, yy1, color='blue', lw=3)
plt.plot(xx, yy1_deriv, color='green', lw=2)

# plt.plot(xx, yy2, color='blue', lw=3)
# slopes = np.linspace(f_1(x_0), 1.8, 4)
# colours = ['g', 'c', 'm', 'y']
# for slope, colour in zip(slopes[:1], colours[:1]):
#    yy_t = (xx - x_0) * slope + f(x_0)
#    plt.plot(xx, yy_t, lw=2, color=colour)



plt.plot([x_0], [y_0], marker='o', color='black', markersize=10)
```
