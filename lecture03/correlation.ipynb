```python
% matplotlib inline
import pylab
pylab.rcParams['figure.figsize'] = (16, 4)

import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.rand(200) * 2 - 1
Y1 = np.random.rand(200) * 2 - 1
Y2 = X + np.random.randn(200) * 0.6
Y3 = -X + np.random.randn(200) * 0.6
Z = np.random.randn(200) > 0
Y4 = (2*Z  -1) * X**2 + np.random.randn(200) * 0.3

fig = plt.figure()
corr1 = 0.
corr2 = np.sqrt(2./3)/np.sqrt(2./3 + 0.6)
corr3 = -corr2
corr4 = 0.
corrs = [corr1, corr2, corr3, corr4]
for (i, Y) in enumerate([Y1, Y2, Y3, Y4]):
    ax = fig.add_subplot('14' + str(i + 1))
    # ax.axison=False
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    ax.set_xlabel('corr = %.1f' % corrs[i], fontsize='25')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.scatter(X, Y)
```
